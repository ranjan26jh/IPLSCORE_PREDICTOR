# Bar Graph

```python
x=[1,2,3,4]
y=[5,6,7,8]


plt.bar(x,y)
plt.show()
```

```python
x=[1,2,3,4]
y=[5,6,7,8]

c=['r','g','b','y']
plt.bar(x,y,color=c)
plt.show()
```

# Scatter Plot

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6]
y=[4,2,3,5,1,8]
plt.title("Satter plot")
plt.xlabel("month")
plt.ylabel("numbers")

plt.scatter(x,y)
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[4,2,3,5,1,8]
plt.title("Satter plot")
plt.xlabel("month")
plt.ylabel("numbers")
c=['red','magenta','green','blue','brown','orange']

plt.scatter(x,y,color=c)
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[4,2,3,5,1,8]
plt.title("Satter plot")
plt.xlabel("month")
plt.ylabel("numbers")
c=['red','magenta','green','blue','brown','orange']

plt.scatter(x,y,color=c,s=150)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

fname='salman.jpeg'

#opening image using pil

image=Image.open(fname).convert('L')

#mapping image to gray scale
plt.imshow(image, cmap='Greys')
plt.title("Salman Khan")
plt.xlabel("length")
plt.ylabel("breadth")
plt.grid()
plt.show()
```

# Pie-Chart

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=['english','hindi','maths','science']
c=['yellow','magenta','brown','aqua']

plt.pie(x,labels=y,colors=c)
plt.legend()
plt.show()
```
